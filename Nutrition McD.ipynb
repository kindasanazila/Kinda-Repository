```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('menu.csv')
df.head()
```

```python
df.describe()
```

```python
print(df.isnull().any())
print('-'*70)

print(df.shape)
print('-'*70)
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(15, 10))
sns.countplot(y='Category', data=df)
```

```python
import warnings
warnings.filterwarnings('ignore')
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
f, ax = plt.subplots(1, 2, figsize=(25, 12))
MeanCalories = {}
MeanCholesterol = {}
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Calories'].mean())
    MeanCalories[category] = mean
pd.Series(MeanCalories).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('RdYlGn',10), ax=ax[0])
ax[0].set_title("Mean Calories of Different Categories")

for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Cholesterol (% Daily Value)'].mean())
    MeanCholesterol[category] = mean
pd.Series(MeanCholesterol).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('RdYlGn',10), ax=ax[1])
ax[1].set_title("Mean Cholesterol of Different Categories")
plt.subplots_adjust(wspace=0.5)
```

```python
f, ax = plt.subplots(1, 2, figsize=(25, 12))
MeanTotalFat = {}
MeanSaturatedFat = {}
for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Total Fat (% Daily Value)'].mean())
    MeanTotalFat[category] = mean
pd.Series(MeanTotalFat).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('inferno',10), ax=ax[0])
ax[0].set_title("Mean Total Fat of Different Categories")

for category in df['Category'].unique():
    mean = int(df[df['Category'] == category]['Saturated Fat (% Daily Value)'].mean())
    MeanSaturatedFat[category] = mean
pd.Series(MeanSaturatedFat).sort_values(ascending=True).plot.barh(width=0.8, color=sns.color_palette('inferno',10), ax=ax[1])
ax[1].set_title("Mean Saturated Fat of Different Categories")
plt.subplots_adjust(wspace=0.5)
```

```python
sns.set(font_scale=1)
f, ax = plt.subplots(3, 3, figsize=(10, 10), sharex=False, sharey=False)

#The distribution of Calories and Colesterol
x = df['Calories'].values
y = df['Cholesterol (% Daily Value)'].values
sns.kdeplot(x, y, ax=ax[0, 0])
sns.rugplot(x, color='g', ax=ax[0, 0])
sns.rugplot(x, vertical=True, ax=ax[0, 0])
ax[0, 0].set(xlim=(-200, 1000), ylim=(-10, 40),  title = 'Calories and Cholesterol')

#The distribution of Total Fat and Colesterol  
x = df['Total Fat (% Daily Value)'].values
y = df['Cholesterol (% Daily Value)'].values
sns.kdeplot(x, y, ax=ax[0, 1])
sns.rugplot(x, color='g', ax=ax[0, 1])
sns.rugplot(x, vertical=True, ax=ax[0, 1])
ax[0, 1].set(xlim=(-20, 70), ylim=(-10, 40), title = 'Total Fat and Cholesterol')
```
